```python
import pandas as pd
```

```python
budgets_df = pd.read_csv('../data/zippedData/tn.movie_budgets.csv.gz', compression = 'gzip')
budgets_df.head()
```

```python
#limit data to movies released after the year 2000
budgets_df['release_year'] = [int(x[-4:]) for x in budgets_df['release_date']]
budgets_df= budgets_df[budgets_df['release_year']>2000]
```

```python
dups_df = budgets_df[budgets_df.duplicated(['movie'], keep = False)]

dups_df
```
