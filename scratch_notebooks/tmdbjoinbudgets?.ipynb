```python
import pandas as pd
import numpy as np
```

```python
tmdb_df = pd.read_csv('../data/zippedData/tmdb.movies.csv.gz', compression = 'gzip')
tmdb_df.head()
```

```python
budgets_df = pd.read_csv('../data/zippedData/tn.movie_budgets.csv.gz', compression = 'gzip')
budgets_df.head()
```

```python
budgets_df['production_budget'] = [int(x.strip('$').replace(',', '')) for x in budgets_df['production_budget']]
lower_budget_df = budgets_df[budgets_df['production_budget'] <= 1500000]
```

```python
lower_budget_df['release_year'] = [int(x[-4:]) for x in lower_budget_df['release_date']]
lower_budget_df.head()
```
